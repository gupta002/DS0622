```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# Load the data and perform necessary preprocessing
# ...

# Create a DataFrame with the movie data
data = {
    'Movie Name': ['The Avengers', 'Jurassic Park', 'Titanic'],
    'Votes': [150000, 100000, 200000],
    'Gross Collection': [3000000, 2000000, 4000000],
    'Star 1': ['Robert Downey Jr.', 'Sam Neill', 'Leonardo DiCaprio']
}

df = pd.DataFrame(data)

# Split the data into features and labels for gross collection prediction
gross_features = df[['Votes', 'Star 1']]
gross_labels = df['Gross Collection']

# Encode categorical column in gross_features
label_encoder = LabelEncoder()
gross_features['Star 1'] = label_encoder.fit_transform(gross_features['Star 1'])

# Split the data into training and testing sets
gross_features_train, gross_features_test, gross_labels_train, gross_labels_test = train_test_split(
    gross_features, gross_labels, test_size=0.2, random_state=42)

# Train the model for gross collection prediction
gross_model = LinearRegression()  # Example model, you can use any suitable regression model
gross_model.fit(gross_features_train, gross_labels_train)

# Create new data for prediction
new_data = {
    'Movie Name': ['Avatar', 'Inception'],
    'Votes': [250000, 180000],
    'Star 1': ['Sam Worthington', 'Leonardo DiCaprio']
}

new_df = pd.DataFrame(new_data)

# Encode categorical column in new_df using label_encoder.transform if label is in label_encoder.classes_,
# otherwise assign a new integer label
new_df['Star 1'] = new_df['Star 1'].apply(
    lambda x: label_encoder.transform([x])[0] if x in label_encoder.classes_ else len(label_encoder.classes_))

# Make predictions on new data
new_gross_predictions = gross_model.predict(new_df[['Votes', 'Star 1']])

# Print evaluation metrics and predictions
gross_mse = mean_squared_error(gross_labels_test, gross_model.predict(gross_features_test))
print("Gross Collection MSE:", gross_mse)
print("New gross collection predictions:", new_gross_predictions)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import joblib

# Create a DataFrame with the movie data
data = {
    'Movie Name': ['The Avengers', 'Jurassic Park', 'Titanic'],
    'Votes': [150000, 100000, 200000],
    'Gross Collection': [3000000, 2000000, 4000000],
    'Star 1': ['Robert Downey Jr.', 'Sam Neill', 'Leonardo DiCaprio']
}

df = pd.DataFrame(data)

# Split the data into features and labels for gross collection prediction
gross_features = df[['Votes', 'Star 1']]
gross_labels = df['Gross Collection']

# Encode categorical column in gross_features
label_encoder = LabelEncoder()
gross_features['Star 1'] = label_encoder.fit_transform(gross_features['Star 1'])

# Split the data into training and testing sets
gross_features_train, gross_features_test, gross_labels_train, gross_labels_test = train_test_split(
    gross_features, gross_labels, test_size=0.2, random_state=42)

# Define the parameter grid for hyperparameter tuning
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5, 10]
}

# Perform hyperparameter tuning using RandomizedSearchCV
rf_model = RandomForestRegressor(random_state=42)
random_search = RandomizedSearchCV(rf_model, param_grid, n_iter=10, scoring='neg_mean_squared_error',
                                   cv=2, random_state=42)
random_search.fit(gross_features_train, gross_labels_train)

# Get the best model from hyperparameter tuning
best_rf_model = random_search.best_estimator_

# Create new data for prediction
new_data = {
    'Movie Name': ['Avatar', 'Inception'],
    'Votes': [250000, 180000],
    'Star 1': ['Sam Worthington', 'Leonardo DiCaprio']
}

new_df = pd.DataFrame(new_data)

# Encode categorical column in new_df using label_encoder.transform if label is in label_encoder.classes_,
# otherwise assign a new integer label
new_df['Star 1'] = new_df['Star 1'].apply(
    lambda x: label_encoder.transform([x])[0] if x in label_encoder.classes_ else len(label_encoder.classes_))

# Make predictions on new data using the best model
new_gross_predictions = best_rf_model.predict(new_df[['Votes', 'Star 1']])

# Print evaluation metrics and predictions
gross_mse = mean_squared_error(gross_labels_test, best_rf_model.predict(gross_features_test))
print("Gross Collection MSE:", gross_mse)
print("New gross collection predictions:", new_gross_predictions)
```

The chosen model is the random forest regressor with the hyperparameters that minimize the mean squared error on the training data during the hyperparameter tuning process.
