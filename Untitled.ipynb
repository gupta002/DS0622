```python
import csv
import requests
from bs4 import BeautifulSoup

# URL to scrape
base_url = "https://www.imdb.com/search/title/?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=f11158cc-b50b-4c4d-b0a2-40b32863395b&pf_rd_r=XZ8X52H1R40B7KG5SNZ9&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1"

# Initialize CSV file 1
csv_file1 = open("movie_details.csv", "w", newline="", encoding="utf-8")
writer1 = csv.writer(csv_file1)
writer1.writerow(["Sno", "Movie Name", "Director 1", "Director 2", "Year", "Rating", "Metascore", "Duration", "Genre 1", "Genre 2", "Genre 3"])

# Initialize CSV file 2
csv_file2 = open("movie_additional_details.csv", "w", newline="", encoding="utf-8")
writer2 = csv.writer(csv_file2)
writer2.writerow(["Movie Name", "Star 1", "Star 2", "Star 3", "Star 4", "Votes", "Genre 1", "Genre 2", "Genre 3", "Gross Collection", "Popularity", "Certification", "Duration"])

page = 1

while True:
    # Create URL for the current page
    url = base_url + "&start=" + str((page - 1) * 50 + 1)

    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.content, "html.parser")
        # Find and extract specific data from the HTML structure
        movies = soup.find_all("div", class_="lister-item-content")

        if not movies:
            # No more movies found, exit the loop
            break

        for movie in movies:
            # Extract movie details
            sno = movie.find("span", class_="lister-item-index").text.strip(".")
            movie_name = movie.find("h3", class_="lister-item-header").find("a").text.strip()

            # Extract director names using href link
            directors = movie.find_all("a", href=lambda href: href and "/name/" in href)
            director_names = [director.text for director in directors]

            # Store director names in separate variables
            director1 = director_names[0] if len(director_names) >= 1 else ""
            director2 = director_names[1] if len(director_names) >= 2 else ""
            director3 = director_names[2] if len(director_names) >= 3 else ""

            year = movie.find("span", class_="lister-item-year").text.strip("()")
            duration_element = movie.find("span", class_="runtime")
            duration = duration_element.text if duration_element else ""
            rating = movie.find("div", class_="inline-block ratings-imdb-rating").strong.text
            metascore = movie.find("div", class_="inline-block ratings-metascore").span.text.strip() if movie.find("div", class_="inline-block ratings-metascore") else ""

            # Extract additional movie details
            stars = movie.find("p", class_="").find_all("a")
            star1 = stars[0].text if len(stars) >= 1 else ""
            star2 = stars[1].text if len(stars) >= 2 else ""
            star3 = stars[2].text if len(stars) >= 3 else ""
            star4 = stars[3].text if len(stars) >= 4 else ""
            votes = movie.find("p", class_="sort-num_votes-visible").find_all("span")[1].text
            genres = movie.find("span", class_="genre").text.strip().split(", ")
            genre1 = genres[0] if len(genres) >= 1 else ""
            genre2 = genres[1] if len(genres) >= 2 else ""
            genre3 = genres[2] if len(genres) >= 3 else ""
            gross_collection = movie.find("p", class_="sort-num_votes-visible").find_all("span")[4].text if len(movie.find("p", class_="sort-num_votes-visible").find_all("span")) >= 5 else ""
            popularity_element = movie.find("span", class_="lister-item-index unbold text-primary")
            popularity = popularity_element.text.strip() if popularity_element else ""
            certification = movie.find("span", class_="certificate").text if movie.find("span", class_="certificate") else ""

            # Write data to CSV file 1
            writer1.writerow([sno, movie_name, director1, director2, year, rating, metascore, duration, genre1, genre2, genre3])

            # Write data to CSV file 2
            writer2.writerow([movie_name, star1, star2, star3, star4, votes, genre1, genre2, genre3, gross_collection, popularity, certification, duration])

            # Print movie details
            print("Movie Name:", movie_name)
            print("Sno:", sno)
            print("Director 1:", director1)
            print("Director 2:", director2)
            print("Director 3:", director3)
            print("Year:", year)
            print("Rating:", rating)
            print("Metascore:", metascore)
            print("Duration:", duration)
            print("Stars:", star1, star2, star3, star4)
            print("Votes:", votes)
            print("Genres:", genre1, genre2, genre3)
            print("Gross Collection:", gross_collection)
            print("Popularity:", popularity)
            print("Certification:", certification)
            print()

        page += 1
    else:
        print("Request failed with status code:", response.status_code)
        break

# Close the CSV files
csv_file1.close()
csv_file2.close()

print("Data has been scraped, stored in CSV files, and printed.")
```

```python
import sqlite3
import csv

# Connect to SQLite database
conn = sqlite3.connect("movies.db")
cursor = conn.cursor()

# Create table 1: movie_details
cursor.execute('''CREATE TABLE movie_details (
                    Sno INTEGER,
                    MovieName TEXT,
                    Director1 TEXT,
                    Director2 TEXT,
                    Year TEXT,
                    Rating REAL,
                    Metascore TEXT,
                    Duration TEXT,
                    Genre1 TEXT,
                    Genre2 TEXT,
                    Genre3 TEXT
                )''')

# Create table 2: movie_additional_details
cursor.execute('''CREATE TABLE movie_additional_details (
                    MovieName TEXT,
                    Star1 TEXT,
                    Star2 TEXT,
                    Star3 TEXT,
                    Star4 TEXT,
                    Votes TEXT,
                    Genre1 TEXT,
                    Genre2 TEXT,
                    Genre3 TEXT,
                    GrossCollection TEXT,
                    Popularity TEXT,
                    Certification TEXT,
                    Duration TEXT
                )''')

# Read and insert data from movie_details.csv
with open('movie_details.csv', 'r', encoding='utf-8') as file:
    csv_reader = csv.reader(file)
    next(csv_reader)  # Skip header row
    for row in csv_reader:
        cursor.execute('INSERT INTO movie_details VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)', row)

# Read and insert data from movie_additional_details.csv
with open('movie_additional_details.csv', 'r', encoding='utf-8') as file:
    csv_reader = csv.reader(file)
    next(csv_reader)  # Skip header row
    for row in csv_reader:
        cursor.execute('INSERT INTO movie_additional_details VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)', row)

# Commit the changes and close the connection
conn.commit()
conn.close()

print("Data has been inserted into SQLite database.")
```
